### 🎯 Ejercicio Práctico: Identificación de KPIs con el Framework **SMART-ER**  
**Objetivo:** Aplicar el marco **SMART-ER** para definir **KPIs relevantes** utilizando el esquema de base de datos de ventas.

---

## 📌 Contexto

Imagina que eres responsable del área comercial en una empresa que vende productos electrónicos. Tienes acceso a las tablas:

- `customers` (clientes)
- `sales` (ventas)
- `products` (productos)
- `employees` (empleados)
- `payments` (pagos)

Tu objetivo es mejorar la rentabilidad y el rendimiento de ventas en el próximo trimestre. Para hacerlo, debes identificar **KPIs alineados con objetivos SMART-ER**.

---

## 🧠 Marco de Trabajo: **SMART-ER**

| Letra | Significado       | Descripción |
|-------|-------------------|-------------|
| S     | Specific          | Claridad sobre lo que se mide |
| M     | Measurable        | Debe poder medirse cuantitativamente |
| A     | Achievable        | Realista y alcanzable |
| R     | Relevant          | Alineado con los objetivos estratégicos |
| T     | Time-bound        | Con un plazo claro |
| E     | Evaluable         | Debe permitir seguimiento periódico |
| R     | Rewarding         | Que motive e impulse acción |

---

## 📊 Ejercicio: Define 2 KPIs usando SMART-ER

### ✏️ Instrucciones:
Para cada KPI, completa los campos según el esquema de base de datos proporcionado. Usa tablas como `sales`, `products`, `customers` y `employees`.

---

### 🔹 KPI 1: Margen promedio por venta

| Criterio      | Respuesta                                                                 |
|---------------|---------------------------------------------------------------------------|
| **Nombre**    | Margen promedio por venta                                                 |
| **Definición**| Promedio del margen bruto por línea de venta: `(priceEach - costo_producto) / priceEach` |
| **S**         | Medir la rentabilidad promedio de cada producto vendido                   |
| **M**         | Se calcula mensualmente con datos de la tabla `sales`                     |
| **A**         | Es alcanzable mediante análisis de precios y costos                       |
| **R**         | Alineado con el objetivo de aumentar la rentabilidad                      |
| **T**         | Evaluar cada mes durante el próximo trimestre                             |
| **E**         | Puede compararse contra metas mensuales                                 |
| **R**         | Permite ajustar estrategias de precios o promociones                     |

---

### 🔹 KPI 2: Número de clientes activos por vendedor

| Criterio      | Respuesta                                                                 |
|---------------|---------------------------------------------------------------------------|
| **Nombre**    | Clientes activos por vendedor                                             |
| **Definición**| Cantidad de clientes únicos atendidos por cada empleado en últimos 90 días |
| **S**         | Identificar qué vendedores están generando más interacción comercial      |
| **M**         | Contar clientes con compras en los últimos 3 meses desde `sales`          |
| **A**         | Basado en historial previo, es alcanzable                                |
| **R**         | Alineado con el objetivo de optimizar el desempeño del equipo comercial   |
| **T**         | Revisar trimestralmente                                                   |
| **E**         | Fácil de medir con consultas SQL                                          |
| **R**         | Motiva a los vendedores a mantener relaciones activas con sus clientes    |

---

## 🛠️ Propuesta de consulta SQL para el KPI 2:

```sql
SELECT 
    s.employeeNumber,
    COUNT(DISTINCT s.customerNumber) AS clientes_activos
FROM sales s
WHERE s.orderDate >= CURRENT_DATE - INTERVAL '90 days'
GROUP BY s.employeeNumber
ORDER BY clientes_activos DESC;
```
