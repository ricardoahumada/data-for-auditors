## Resumen de Hiperparmetros

| **Algoritmo**                     | **Hiperparmetro**                  | **Descripcin**                                                                                     | **Valores Tpicos**                                                                 |
|-----------------------------------|-------------------------------------|-----------------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------|
| **K-Means**                       | `n_clusters`                       | Nmero de clsteres a formar.                                                                       | Entero positivo (por ejemplo, 2, 3, 5, 10).                                         |
|                                   | `init`                             | Mtodo de inicializacin de los centroides.                                                         | `'k-means++'`, `'random'`.                                                          |
|                                   | `max_iter`                         | Nmero mximo de iteraciones para converger.                                                        | Entero positivo (por ejemplo, 100, 300).                                            |
|                                   | `random_state`                     | Semilla para reproducibilidad.                                                                      | Entero positivo o `None`.                                                           |
| **DBSCAN**                        | `eps`                              | Radio mximo de vecindad para considerar puntos como vecinos.                                       | Flotante positivo (por ejemplo, 0.1, 0.5, 1.0).                                     |
|                                   | `min_samples`                      | Nmero mnimo de puntos en una vecindad para formar un clster.                                     | Entero positivo (por ejemplo, 2, 5, 10).                                            |
|                                   | `metric`                           | Mtrica de distancia utilizada para calcular la cercana entre puntos.                              | `'euclidean'`, `'manhattan'`, `'cosine'`.                                           |
| **Hierarchical Clustering**       | `n_clusters`                       | Nmero de clsteres finales.                                                                        | Entero positivo (por ejemplo, 2, 3, 5).                                             |
|                                   | `affinity`                         | Mtrica de distancia utilizada para calcular la similitud entre puntos.                             | `'euclidean'`, `'manhattan'`, `'cosine'`.                                           |
|                                   | `linkage`                          | Mtodo para combinar clsteres en clustering jerrquico.                                            | `'ward'`, `'complete'`, `'average'`, `'single'`.                                    |
| **Gaussian Mixture Model (GMM)**  | `n_components`                     | Nmero de componentes (distribuciones gaussianas) en la mezcla.                                      | Entero positivo (por ejemplo, 2, 3, 5).                                             |
|                                   | `covariance_type`                  | Tipo de matriz de covarianza para las distribuciones gaussianas.                                     | `'full'`, `'tied'`, `'diag'`, `'spherical'`.                                        |
|                                   | `random_state`                     | Semilla para reproducibilidad.                                                                      | Entero positivo o `None`.                                                           |
| **Principal Component Analysis (PCA)** | `n_components`                 | Nmero de componentes principales a conservar.                                                      | Entero positivo o flotante (por ejemplo, 2, 0.95 para varianza explicada).          |
|                                   | `svd_solver`                       | Mtodo computacional para realizar la descomposicin SVD.                                           | `'auto'`, `'full'`, `'arpack'`, `'randomized'`.                                     |
| **t-SNE**                         | `n_components`                     | Nmero de dimensiones del espacio reducido.                                                         | Entero positivo (por ejemplo, 2, 3).                                                |
|                                   | `perplexity`                       | Balance entre estructura local y global en los datos.                                               | Flotante positivo (por ejemplo, 5, 30, 50).                                         |
|                                   | `learning_rate`                    | Tasa de aprendizaje para optimizar la distribucin de los puntos.                                   | Flotante positivo (por ejemplo, 10, 200, 1000).                                     |
| **UMAP**                          | `n_components`                     | Nmero de dimensiones del espacio reducido.                                                         | Entero positivo (por ejemplo, 2, 3).                                                |
|                                   | `n_neighbors`                      | Nmero de vecinos considerados para preservar la estructura local.                                  | Entero positivo (por ejemplo, 5, 10, 30).                                           |
|                                   | `min_dist`                         | Distancia mnima entre puntos en el espacio reducido.                                               | Flotante positivo (por ejemplo, 0.1, 0.5).                                          |
| **Non-Negative Matrix Factorization (NMF)** | `n_components`             | Nmero de componentes no negativos en la descomposicin.                                            | Entero positivo (por ejemplo, 2, 5, 10).                                            |
|                                   | `init`                             | Mtodo de inicializacin de los factores.                                                           | `'random'`, `'nndsvd'`, `'nndsvda'`, `'nndsvdar'`.                                  |
|                                   | `solver`                           | Algoritmo utilizado para resolver la descomposicin.                                                | `'cd'` (coordenadas descendentes), `'mu'` (multiplicative update).                   |
| **Latent Dirichlet Allocation (LDA)** | `n_components`                 | Nmero de temas latentes a descubrir.                                                               | Entero positivo (por ejemplo, 5, 10, 20).                                           |
|                                   | `learning_method`                  | Mtodo de aprendizaje para estimar los parmetros del modelo.                                       | `'batch'`, `'online'`.                                                              |
|                                   | `random_state`                     | Semilla para reproducibilidad.                                                                      | Entero positivo o `None`.                                                           |
| **Apriori**                       | `min_support`                      | Umbral mnimo de soporte para considerar un conjunto de elementos frecuente.                        | Flotante entre 0 y 1 (por ejemplo, 0.1, 0.5).                                       |
